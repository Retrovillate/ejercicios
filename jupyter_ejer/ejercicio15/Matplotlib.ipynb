```python

"""
Por defecto se crean los graficos fuera de la celdadel jupyter
y se coloca % para indicar la celda
"""
# libreria para graficos en 2D
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generando grafica con un solo array

array = np.array([1,2,3,4,5,6])
```

```python
# graficas en 2D
plt.plot(array)
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# generando arrays para cada eje

arrayX = np.array([1,2,3,4,5,6])
arrayY = np.array([6,5,4,3,2,1])

plt.plot(arrayX, arrayY)
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# graficas con configuracion 

# puntos rojos
plt.plot(arrayX, arrayY, 'pr')
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# graficas con configuracion 

# puntos azules
plt.plot(arrayX, arrayY, 'pb')
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# graficas con configuracion 

# tamaño de la linea
plt.plot(arrayX, arrayY, linewidth = 3)
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# graficas con configuracion 

# tamaño de la linea y a rayas
plt.plot(arrayX, arrayY, '--', linewidth = 3)
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# arrays multiples

array2x = np.random.randint(low=1, high=100, size=10)
array2y = np.random.randint(low=1, high=100, size=10)

plt.plot(arrayX, arrayY, 'vg')
plt.ylabel("Eje y")
plt.xlabel("Eje x")

plt.plot(array2x, array2y, 'co')

plt.show()
```

```python
# histogramas

plt.hist(array2y)
plt.show()
```

```python
valores = np.random.randn(1000)
plt.hist(valores)
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
#ajustar la orientacion del hitograma

valores = np.random.randn(1000)
plt.hist(valores, orientation= 'horizontal')
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
#ajustar la orientacion el color del histograma

valores = np.random.randn(1000)
plt.hist(valores, orientation= 'horizontal', color='red')
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
#ajustar al histograma una distribucion por rangos

valores = np.random.randn(1000)
plt.hist(valores, range(-1,1))
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# distribucion del histograma en 2D

valores = np.random.randn(1000)
valores2 = np.random.randn(1000)
plt.hist2d(valores, valores2)
plt.ylabel("Eje y")
plt.xlabel("Eje x")
plt.show()
```

```python
# generando grafica de barras

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

values_x = ['a','b','c','d','e']
values_y = [1,5,6,7,8]

ax.bar(values_x, values_y)

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_data = 4
values_1 = (90, 55, 40, 65)
values_2 = (85, 62, 54, 20)

# configuramos el plot
fig, ax = plt.subplots()
index = np.arange(n_data)
bar_width = 0.35

# configuramos las barras del primer conjunto de valores
rects1 = plt.bar(index,values_1, bar_width, color = 'r', label = 'Valores 1')

# configuramos las barras del segundo conjunto de valores
rects2 = plt.bar(index+bar_width, values_2, bar_width, color = 'g', label = 'Valores 2')

# Insertamos los titulos y la leyenda del grafico

plt.xlabel("Etiquetas")
plt.ylabel("Valores")
plt.title('Titulo del grafico')
plt.xticks(index+bar_width, ('A','B','C','D'))
plt.legend()

# Pintamos el grafico y lo mostramos
plt.tight_layout()
plt.show()
```

```python
# graficos circulares

values = np.array([0.1,0.5,0.3,0.1])
labels = ['value 1','value 2' ,'value 3','value 4']
plt.pie(values, labels =labels )
plt.show()
```

```python
# graficos circulares reduciendo tamaño

values = np.array([0.1,0.5,0.3,0.1])
labels = ['value 1','value 2' ,'value 3','value 4']
plt.pie(values, labels =labels, radius = 0.5)
plt.show()
```

```python
# graficos circulares ampliando tamaño

values = np.array([0.1,0.5,0.3,0.1])
labels = ['value 1','value 2' ,'value 3','value 4']
plt.pie(values, labels =labels, radius = 2)
plt.show()
```

```python
# incluir varias graficas en un solo marco

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

figure = plt.gcf() # obtenemos la Figura donde se pintaran las graficas
figure.set_size_inches(15, 5) # Configuramos el tamaño en pulgadas

plt.subplot(131) # Fílas 1 - Columnas 3 - Posicion del grafico 1

values = np.array([0.2, 0.10, 0.15, 0.05, 0.16, 0.34])
plt.pie(values)

plt.subplot(132) # Filas 1 - Columnas 3 - Posicion 2

valoresX = np.random.randn(1000)
valoresY = np.random.randn(1000)
plt.hist2d(valoresX, valoresY)

plt.subplot(133)

values_X1 = np.random.randint(low=1, high=100, size=10)
values_Y1 = np.random.randint(low=1, high=100, size=10)
plt.plot(values_X1, values_Y1, 'co')

values_X2 = np.random.randint(low=1, high=100, size=10)
values_Y2 = np.random.randint(low=1, high=100, size=10)
plt.plot(values_X2, values_Y2, 'vg')

plt.show()
```
